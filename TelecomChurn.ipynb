```python
# Columns with 'arpu' information 
arpu_cols = train_df.columns[train_df.columns.str.contains('arpu_')]
print(arpu_cols)

# Replacing NaN values in ARPU related columns with 0 in the training set
train_df[arpu_cols] = train_df[arpu_cols].fillna(0)
```

```python
#Calculating the ARPU average to filter out high value customers
train_df['arpu_avg'] = train_df[arpu_cols].mean(axis=1)
```

```python
# Filter high value customers in the new dataframe with high ARPU average where ARPU avg is greater than 0.5 quartile
train_df = train_df[(train_df['arpu_avg'] > train_df['arpu_avg'].quantile(0.5))]
print(train_df.shape)
```

```python
# create column name list by types of columns
id_cols = ['mobile_number', 'circle_id']

date_cols = ['last_date_of_month_6',
             'last_date_of_month_7',
             'last_date_of_month_8',
             'date_of_last_rech_6',
             'date_of_last_rech_7',
             'date_of_last_rech_8',
             'date_of_last_rech_data_6',
             'date_of_last_rech_data_7',
             'date_of_last_rech_data_8',
            ]

cat_cols =  ['night_pck_user_6',
             'night_pck_user_7',
             'night_pck_user_8',
             'fb_user_6',
             'fb_user_7',
             'fb_user_8',
            ]

num_cols = [col for col in train_df.columns if col not in id_cols + date_cols + cat_cols]
```

```python
# print the number of columns in each list
print('id_cols: %d' % len(id_cols))
print('date_cols: %d' % len(date_cols))
print('cat_cols: %d' % len(cat_cols))
print('num_cols: %d' % len(num_cols))
print(train_df.shape)
```

```python
#Drop data columns as they make no impact
train_df = train_df.drop(columns=date_cols, axis=1)
print(train_df.shape)
```

```python
# look at missing value ratio in each column
train_df.isnull().sum()/train_df.shape[0]*100
```

```python
#Remove all attributes with missing values greater than 50%
THRESHOLD_MISS = 0.5
include_cols = list(train_df.apply(lambda col: True if col.isnull().sum()/train_df.shape[0] < THRESHOLD_MISS else False))
train_df = train_df.loc[:, include_cols]
print(train_df.shape)
```
